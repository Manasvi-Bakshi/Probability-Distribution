Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from mpl_toolkits.mplot3d import Axes3D
```

## Univariate Regression

```python
# Create dataset: y = 5 * x - 2 + noise
np.random.seed(0)
X_univariate = np.random.rand(100, 1) * 10  # Feature in range [0, 10]
y_univariate = 5 * X_univariate - 2 + np.random.randn(100, 1)

# Split data
X_train_univariate, X_test_univariate, y_train_univariate, y_test_univariate = train_test_split(X_univariate, y_univariate, test_size=0.2, random_state=0)

# Train model
model_univariate = LinearRegression()
model_univariate.fit(X_train_univariate, y_train_univariate)

# Predictions and evaluation
y_pred_univariate = model_univariate.predict(X_test_univariate)
mse_univariate = mean_squared_error(y_test_univariate, y_pred_univariate)
print(f'Mean Squared Error (Univariate): {mse_univariate}')

# Plot results
plt.scatter(X_test_univariate, y_test_univariate, color='blue', label='Actual')
plt.scatter(X_test_univariate, y_pred_univariate, color='red', label='Predicted')
plt.plot(X_test_univariate, y_pred_univariate, color='green')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Univariate Regression')
plt.legend()
plt.show()
```

## Multivariate Regression


```python
# Create dataset: y = 4 * x1 + 2 * x2 + noise
X1 = np.random.rand(100, 1) * 10
X2 = np.random.rand(100, 1) * 5
y_multivariate = 4 * X1 + 2 * X2 + np.random.randn(100, 1)

# Split data
X_multi = np.hstack((X1, X2))
y_multi = y_multivariate
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.2, random_state=0)

# Train model
model_multivariate = LinearRegression()
model_multivariate.fit(X_train_multi, y_train_multi)

# Predictions and evaluation
y_pred_multi = model_multivariate.predict(X_test_multi)
mse_multi = mean_squared_error(y_test_multi, y_pred_multi)
print(f'Mean Squared Error (Multivariate): {mse_multi}')

# 3D Plot for Multivariate Regression
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_test_multi[:, 0], X_test_multi[:, 1], y_test_multi, color='blue', label='Actual')
ax.scatter(X_test_multi[:, 0], X_test_multi[:, 1], y_pred_multi, color='red', label='Predicted')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')
ax.set_title('Multivariate Regression')
ax.legend()
plt.show()
```
