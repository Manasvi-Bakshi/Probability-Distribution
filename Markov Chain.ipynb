```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Data

```python
# Define states and transition probabilities
states = ["A", "B", "C", "D"]
transitions = np.array([
    [0.3, 0.4, 0.2, 0.1],  # A
    [0.2, 0.3, 0.3, 0.2],  # B
    [0.3, 0.3, 0.2, 0.2],  # C
    [0.1, 0.2, 0.4, 0.3]   # D
])

def simulate_chain(transitions, states, start, steps):
    current = states.index(start)
    sequence = [start]

    for _ in range(steps):
        next_state = np.random.choice(states, p=transitions[current])
        sequence.append(next_state)
        current = states.index(next_state)

    return sequence

# Parameters
start = "A"
steps = 50

# Run simulation
sequence = simulate_chain(transitions, states, start, steps)

# Plot results
plt.figure(figsize=(12, 6))
plt.plot(sequence, marker="o", linestyle="--")
plt.title("State Transitions in Markov Chain")
plt.xlabel("Steps")
plt.ylabel("State")
plt.show()

# Visualize transition probabilities
plt.figure(figsize=(8, 6))
sns.heatmap(transitions, annot=True, cmap="Blues", xticklabels=states, yticklabels=states)
plt.title("Transition Matrix")
plt.xlabel("Next State")
plt.ylabel("Current State")
plt.show()
```
